```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('seaborn')
```

```python
#  Data Preparation
# reading data 
dfx = pd.read_csv('./Training Data/Logistic_X_Train.csv')
dfy = pd.read_csv('./Training Data/Logistic_Y_Train.csv')

#convert into numpy array
x = np.asarray(dfx)
y = np.asarray(dfy)

print(x.shape)
print(y.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.scatter3D(x[:,1],x[:,2],x[:,0],  cmap='hsv')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x,y)
```

```python
test_data = pd.read_csv('./Testing Data/Logistic_X_Test.csv')
test_data = np.asarray(test_data)
output = model.predict(test_data)
# output
print(test_data[0])
model.predict(test_data[0].reshape((1,-1)))
```

```python
model.score(x,y)
```
