```python
def create_plot(ptype):
    x=np.arange(-10,10,0.01)
    if(ptype=='linear'):
        y=x
    elif(ptype=='quadratic'):
        y=x**2
    elif(ptype=='cubic'):
        y=x**3
    else:
        y==x**4
    return (x,y)

fig=plt.figure()
```
